```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset from the CSV file
data = pd.read_csv('/content/India_covid_data.csv')

# Convert 'Date' column to datetime format
data['Date'] = pd.to_datetime(data['Date'])

# Check for missing values
print(data.isnull().sum())

# Drop rows with missing values (if any)
data = data.dropna()

# Filter data for India
india_data = data[data['Country'] == 'India']

# Group data by date and sum total cases
india_cases = india_data.groupby('Date')['Cases'].sum().reset_index()

# Plotting total cases over time for India
plt.figure(figsize=(14, 7))
plt.plot(india_cases['Date'], india_cases['Cases'], color='blue', label='Total Cases')
plt.title('Total Covid-19 Cases Over Time in India')
plt.xlabel('Date')
plt.ylabel('Total Cases')
plt.xticks(rotation=45)
plt.legend()
plt.show()

# Analyze correlation between vaccinations and cases
vaccination_cases = india_data[['Date', 'Cases', 'Vaccinated']].dropna()

# Scatter plot
plt.figure(figsize=(14, 7))
sns.scatterplot(data=vaccination_cases, x='Vaccinated', y='Cases', color='blue')
plt.title("Total Cases vs Total Vaccinations in India")
plt.xlabel("Total Vaccinations")
plt.ylabel("Total Cases")
plt.xticks(rotation=45)
plt.show()
```
